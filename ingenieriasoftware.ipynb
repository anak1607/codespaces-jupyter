* Además de las pymes, los stakeholders incluyen:

## SOLUCION REQUERIDA
## Partes interesadas identificadas
* Pymes latinoamericanas pequeñas: principales usuarios, buscan simplicidad y bajo costo.
* Dueños/gerentes de negocios pequeños: necesitan control rápido de clientes y ventas sin procesos complejos.
* Equipos de ventas y atención al cliente: requieren organización de conversaciones y recordatorios.
* Proveedores de servicios financieros y facturación electrónica: actores clave para integración local.
* Inversionistas en SaaS regionales: interesados en soluciones escalables y adaptadas al mercado.
* Clientes finales de las pymes: beneficiados indirectamente por una mejor gestión de comunicación.
## Necesidades de las partes interesadas que generan la oportunidad
* Simplicidad operativa: evitar configuraciones técnicas y procesos largos.
* Integración directa con WhatsApp Business: canal de ventas predominante en la región.
* Facturación básica y pagos locales: cumplir con normativas y hábitos de consumo.
* Visualización clara del embudo de ventas: estados simples (Nuevo, En conversación, Cotización, Cerrado).
* Accesibilidad económica: precios ajustados a microempresas.
* Rapidez de uso: onboarding inmediato, sin capacitación extensa.

## Problemas subyacentes y sus causas
* Complejidad de CRMs globales: diseñados para empresas grandes, con exceso de funciones.
* Baja adopción tecnológica en pymes: falta de tiempo y recursos para aprender herramientas sofisticadas.
* Fragmentación de procesos: ventas por WhatsApp, facturación aparte, recordatorios manuales.
* Costos elevados de soluciones internacionales: inaccesibles para microempresas.
* Desajuste cultural y lingüístico: interfaces en inglés, sin adaptación a pagos locales.

## Confirmación de la necesidad de una solución sistémica
* Los problemas no se resuelven con herramientas aisladas (ej. solo facturación o solo chat).
* Se requiere una plataforma integrada que conecte comunicación, ventas y facturación en un flujo único.
* La solución debe ser sistémica porque involucra varios actores (pymes, clientes, proveedores de servicios financieros) y procesos interdependientes.

# VALOR ESTABLECIDO

## Valor cuantificado en términos absolutos o de rentabilidad/ahorro
* Ahorro de tiempo: si una pyme dedica actualmente 5–10 horas semanales a organizar clientes manualmente, el CRM podría reducir ese esfuerzo en un 60–70%.
* Ahorro económico: evitar el pago de soluciones globales (USD 50–100 por usuario/mes) y ofrecer una alternativa local (ej. USD 10–15 por usuario/mes).
* Rentabilidad: al mejorar el seguimiento de clientes, se estima un aumento de cierre de ventas del 10–20% anual.
* Ejemplo: una pyme con ingresos de USD 100,000/año podría ganar entre USD 10,000–20,000 adicionales gracias a mejor gestión.

* Dueños de negocios: mayor control de ventas y clientes sin necesidad de conocimientos técnicos.
* Equipos de ventas: organización clara de conversaciones y recordatorios automáticos.
* Clientes finales: respuestas más rápidas y seguimiento más profesional.
* Inversionistas: acceso a un mercado amplio y poco atendido, con potencial de escalabilidad regional.
